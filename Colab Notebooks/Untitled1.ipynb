```python
import os, glob

# ── Auto-detect path CSV La Liga ──────────────────────────
def find_file(patterns):
    for pat in patterns:
        found = glob.glob(pat, recursive=True)
        if found:
            return found[0]
    return None

LALIGA_CSV  = find_file([
    "/content/laliga2526/*.csv",
    "/content/laliga2526/**/*.csv",
    "/content/**/*laliga*stats*.csv",
    "/content/**/*la-liga*.csv",
])

FIFA_CSV    = find_file([
    "/content/fifa25/male_players.csv",
    "/content/fifa25/**/*.csv",
])

PLAYERS_CSV = find_file([
    "/content/players2526/players_data-2025_2026.csv",
    "/content/players2526/**/*.csv",
])

INJURY_CSV  = find_file([
    "/content/injuries/full_dataset_thesis*.csv",
    "/content/injuries/**/*.csv",
])

print("📁 Path yang ditemukan:")
print(f"  La Liga : {LALIGA_CSV}")
print(f"  FIFA25  : {FIFA_CSV}")
print(f"  Players : {PLAYERS_CSV}")
print(f"  Injuries: {INJURY_CSV}")
```
